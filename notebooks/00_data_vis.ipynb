```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from IPython.display import display, Markdown
```

```python
DATA_DIR = Path("../data")
OUTPUTS_DIR = Path("../outputs/")
N_SAMPLES_PER_TABLE = 10

assert DATA_DIR.exists(), "data/ directory not found"
os.makedirs(OUTPUTS_DIR, exist_ok=True)
```

```python
# List all parquet files
parquet_files = sorted([f for f in os.listdir(DATA_DIR) if f.endswith(".parquet")])
print(f"Found {len(parquet_files)} parquet tables:\n" + "\n".join(f"  - {f}" for f in parquet_files))
```

```python
markdown_str = ""

for i, f in enumerate(parquet_files):
    extless_filename = f.split(".")[0]
    markdown_str += f"## {i:02d} - Table `{f}`\n\n"
    full_filename = os.path.join(DATA_DIR, f)
    df = pd.read_parquet(full_filename)

    markdown_str += f"### Table `{extless_filename}` sample\n\n"
    markdown_str += df.sample(n=min(N_SAMPLES_PER_TABLE, len(df.index))).to_markdown() + "\n\n"

    markdown_str += f"### Table `{extless_filename}` shape\n\n"
    markdown_str += f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns\n\n\n"
    markdown_str += f"### Table `{extless_filename}` data types\n\n"
    markdown_str += df.dtypes.to_markdown() + "\n\n"
    nulls = df.isnull().sum().to_markdown()
    markdown_str += f"### Table `{extless_filename}` null counts\n\n"
    markdown_str += nulls + "\n\n"

    markdown_str += "\n---\n\n\n"

display(Markdown(markdown_str))
```

```python
# write the markdown_str to a file
with open(os.path.join(OUTPUTS_DIR, "00_data_description.md"), "w", encoding="utf-8") as md_file:
    md_file.write(markdown_str)
```
